```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
from sklearn.cluster import KMeans
```

# 1.Chargement du jeu de données

```python
df=pd.read_csv("data/CC GENERAL.csv", encoding="iso-8859-1")
df
```

```python
df.isnull().sum()
```

```python
df['MINIMUM_PAYMENTS'].fillna(df['MINIMUM_PAYMENTS'].mode()[0], inplace=True)
```

```python
df['CREDIT_LIMIT'].fillna(df['CREDIT_LIMIT'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df=df.drop("CUST_ID", axis=1)
```

```python
df["CREDIT_LIMIT"].describe()
```

```python
df
```

# 2. Clustering hierarchique

```python
model=AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='complete')
clust_labels=model.fit_predict(df)
```

```python
clust_labels
```

```python
agglomerative=pd.DataFrame(clust_labels)
agglomerative
```

# 3. Visualisation

```python
fig =plt.figure()
#plt.axis([0, 20000, 0, 50000])
ax = fig.add_subplot(111)
scatter = ax.scatter (df['BALANCE'] , df['PURCHASES'] , c= agglomerative[0], s=50)
ax.set_title("Agglomerative Clustering")
ax.set_xlabel("BALANCE")
ax.set_ylabel("PURCHASES")
plt.colorbar(scatter)
```

# 4. Dendrogramme et clustering k-means

```python

plt.figure(figsize=(10,7))
plt.title("Customer Dendrograms")
dend=shc.dendrogram(shc.linkage(df, method="complete"))
```

```python
kmeans=KMeans(n_clusters=5, random_state=0)
kmeans.fit(df)
```

```python
labels=pd.DataFrame(kmeans.labels_)
labels
```

```python
x=df.iloc[:,[0,2]]
kmeans.predict(x)
print(kmeans.cluster_centers_)
```

```python
x=df.iloc[:,[0,2]]
y_kmeans=kmeans.fit_predict(x)
```

```python
y_kmeans
```

```python
x
```

```python
plt.scatter(df["BALANCE"][labels[0] == 0],          
            df["PURCHASES"][labels[0] == 0],s=80,c='magenta',label='Careful')
plt.scatter(df["BALANCE"][labels[0] == 1],
           df["PURCHASES"][labels[0] == 1],s=80,c='yellow',label='Standard')
plt.scatter(df["BALANCE"][labels[0] == 2],
           df["PURCHASES"][labels[0] == 2],s=80,c='green',label='Target')
plt.scatter(df["BALANCE"][labels[0] == 3], 
           df["PURCHASES"][labels[0] == 3],s=80,c='cyan',label='Careless')
plt.scatter(df["BALANCE"][labels[0] == 4], 
           df["PURCHASES"][labels[0] == 4],s=80,c='burlywood',label='Sensible')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label = 'Centroids')
plt.title('Clusters of Customers')
plt.xlabel('BALANCE')
plt.ylabel('PURCHASES')
plt.legend()
plt.show()
```

# 5. Meilleure valeur de k

```python
sum_of_squared_distances = []
k=range(1,15)
for k in k:
    km=KMeans(n_clusters=k)
    km=km.fit(df)
    sum_of_squared_distances.append(km.inertia_)
```

```python
k=range(1,15)
plt.plot(k, sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal K')
plt.show()
```
